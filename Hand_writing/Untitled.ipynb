```python
import os, io
from google.cloud import vision
from google.cloud.vision import types
import pandas as pd
```

```python
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = r"/Users/huybui/Desktop/Google_Cloud_API/key.json"
client = vision.ImageAnnotatorClient()

FILE = 'ERM.jpg'
FOLDER = r'/Users/huybui/Desktop/Google_Cloud_API/Code/Hand_writing'
PATH=os.path.join(FOLDER,FILE)
```

```python
with io.open(PATH, 'rb') as image_file:
    content = image_file.read()
```

```python
image=vision.types.Image(content=content)
response=client.document_text_detection(image=image)
```

```python
response
```

```python
description=response.text_annotations[0].description
```

```python
print(description)
```

```python
pages=response.full_text_annotation.pages
for page in pages:
    for block in page.blocks:
        print('Block confidence:',block.confidence)
        for word in block.paragraphs:
            word_text="".join([symbol.text for symbol in word.symbols])
            print("Word text: {} confidence: {}".format(word_text,word.confidence))
            
            for symbol in word.symbols:
                print('\t Symbol:{} confidence: {}'.format(symbol.text, symbol.confidence))
```
